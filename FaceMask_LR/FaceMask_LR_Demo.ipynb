##**Import thư viện cần dùng**

```
from google.colab.patches import cv2_imshow
import pickle
import cv2
import numpy as np
```

##**Load model**

```
# Load model
file_model =  open('/content/drive/My Drive/CS114/FaceMask/Model_LR/Model_LR.pkl', 'rb') 
Model_LR = pickle.load(file_model)
```

##**Tiến hành demo**


```
# Demo
img = cv2.imread("/content/example (4).jpg")

cv2_imshow(cv2.resize(img,(224,224)))
img = cv2.resize(img,(32,32)).flatten()
img = np.array(img).reshape(1,-1)/255

result = Model_LR.predict(img)

if (result == 1):
  print("----------------------------Mask----------------------------")
else: 
  print("----------------------------Without_Mask----------------------------")
```
